# **Section I: Introduction**

The Fashion-MNIST Clothing Classification Project is a more realistic and challenging extension of the classic MNIST project. Instead of handwritten digits, the dataset consists of grayscale images of **fashion items** such as shirts, sneakers, bags, and coats. The goal is to train a Convolutional Neural Network (CNN) capable of recognizing and classifying these 10 categories of clothing items.

This dataset is more challenging than MNIST because:

- Some classes look visually similar (e.g., *Shirt* vs *T-shirt*, *Sneaker* vs *Ankle boot*)
- Clothing textures and shapes vary significantly  
- Fashion items have less contrast than digits  

This project shows how CNNs learn advanced visual patterns and how techniques such as **batch normalization, deeper architectures, and data augmentation** improve model performance.

The main objective is to build and evaluate a powerful CNN that performs accurate classification and to analyze errors to understand which classes are most difficult for the model.

### **Input:**  
A 28×28 grayscale image of a clothing item.

### **Output:**  
A label between 0–9 representing the clothing category.

Supervised Learning → Image Classification

# **Section IV: Data Preprocessing**

To prepare the dataset for training:

# **Data Augmentation**
To improve generalization and reduce overfitting, we apply **data augmentation**, including:

- **Random rotations (±15°):** Teaches the model to recognize items even if slightly tilted.  
- **Horizontal flips:** Helps with symmetrical items such as shirts and shoes.  
- **Random zoom:** Handles scale variations in images.  
- **Width and height shifts:** Ensures the model can recognize items even if not perfectly centered.  

**Benefit:** Data augmentation simulates real-world variations in fashion images, making the model more robust.

# **Section V: Model Architecture**
The CNN model for Fashion-MNIST uses a deeper structure than MNIST due to higher complexity.

## **Layer-by-Layer Architecture**
1. **Conv2D (32 filters, 3×3 kernel)**  
   Extracts edges and simple patterns  
   **ReLU activation**

2. **Batch Normalization**

3. **MaxPooling (2×2)**  
   Reduces image size and computation

4. **Conv2D (64 filters, 3×3 kernel)**  
   Learns complex features (textures, curves)  
   **ReLU activation**

5. **Batch Normalization**

6. **MaxPooling (2×2)**

7. **Conv2D (128 filters, 3×3 kernel)**  
   Extracts higher-level patterns  
   **ReLU activation**

8. **Flatten Layer**  
   Converts 2D feature maps to a 1D vector

9. **Dense (128 units, ReLU)**  
   Learns high-level representations

10. **Dropout (0.3–0.5)**  
    Prevents overfitting

11. **Dense (10 units, Softmax)**  
    Outputs probabilities for each class

## **Why CNN?**
- Automatically learns spatial features  
- Translation invariance improves recognition  
- Fewer parameters than fully connected networks

````python
from tensorflow.keras.datasets import fashion_mnist

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

print("Training set:", x_train.shape, y_train.shape)
print("Test set:", x_test.shape, y_test.shape)
````

# lets visualize our dataset

````python
import matplotlib.pyplot as plt
import numpy as np

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize=(12, 6))
for i in range(10):
    idx = np.where(y_train == i)[0][0]
    plt.subplot(2, 5, i+1)
    plt.imshow(x_train[idx], cmap='gray')
    plt.title(class_names[i])
    plt.axis('off')
plt.show()
````

# **NORMALIZATION:**

````python
x_train = x_train / 255.0
x_test = x_test / 255.0

print("Min pixel value:", x_train.min())
print("Max pixel value:", x_train.max())
````

# **Reshaping:**

````python
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

print("X Training set:", x_train.shape)
print("X Test set:", x_test.shape)
````

# **PREPARE LABELS:**

````python
from tensorflow.keras.utils import to_categorical

y_train_cat = to_categorical(y_train, 10)
y_test_cat= to_categorical(y_test, 10)

print("Y Training set: ", y_train_cat.shape)
print("Y Test set: ", y_test_cat.shape)
````

# **Building the CNN model**

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

model = Sequential()

model.add(Conv2D(32, kernel_size=3, activation='relu', input_shape=(28,28,1)))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64, kernel_size=3, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(2,2))

model.add(Conv2D(128, kernel_size=3, activation='relu'))
model.add(BatchNormalization())

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(10, activation='softmax'))

model.summary()
````

# **COMPILE:**

````python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
````

# **Implement data augmentation: random rotations (±15°), horizontal flips, random cropping**

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=15,
    horizontal_flip=True
)

datagen.fit(x_train)
````

````python
import matplotlib.pyplot as plt
import numpy as np

aug_iter = datagen.flow(x_train[:10], y_train_cat[:10], batch_size=10)
aug_images, aug_labels = next(aug_iter)

plt.figure(figsize=(12,6))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(aug_images[i].reshape(28,28), cmap='gray')
    plt.axis('off')
plt.show()
````

````python
augmented_batch = next(aug_iter)
aug_images = augmented_batch[0]

plt.figure(figsize=(12, 6))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(aug_images[i].reshape(28, 28), cmap='gray')
    plt.axis('off')
plt.show()
````

````python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

datagen.fit(x_train)
````

````python
history = model.fit(
    datagen.flow(x_train, y_train_cat, batch_size=32),
    epochs=20,
    validation_data=(x_test, y_test_cat)
)
````
